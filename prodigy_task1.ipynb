```python
import zipfile
import pandas as pd

# Unzip the dataset
with zipfile.ZipFile('API_SP.POP.TOTL_DS2_en_csv_v2_1114030.zip', 'r') as zip_ref:
    zip_ref.extractall('world_population')

# Load the CSV file
csv_file_path = 'world_population/API_SP.POP.TOTL_DS2_en_csv_v2_1114030.csv'
data = pd.read_csv(csv_file_path, skiprows=4)  # Skipping the first 4 rows
```

```python
# Extract population data for the year 2023
population_2023 = data[['Country Name', '2023']].dropna()

# Sort the data by population in descending order
population_2023_sorted = population_2023.sort_values(by='2023', ascending=False)
```

```python
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(population_2023_sorted['Country Name'][:10], population_2023_sorted['2023'][:10])
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Top 10 Countries by Population in 2023')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt

# Unzip the dataset
with zipfile.ZipFile('API_SP.POP.TOTL_DS2_en_csv_v2_1114030.zip', 'r') as zip_ref:
    zip_ref.extractall('world_population')

# Load the CSV file
csv_file_path = 'world_population/API_SP.POP.TOTL_DS2_en_csv_v2_1114030.csv'
data = pd.read_csv(csv_file_path, skiprows=4)  # Skipping the first 4 rows

# Inspect the data
print(data.head())

# Extract population data for the year 2023
population_2023 = data[['Country Name', '2023']].dropna()

# Sort the data by population in descending order
population_2023_sorted = population_2023.sort_values(by='2023', ascending=False)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(population_2023_sorted['Country Name'][:10], population_2023_sorted['2023'][:10])
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Top 10 Countries by Population in 2023')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
